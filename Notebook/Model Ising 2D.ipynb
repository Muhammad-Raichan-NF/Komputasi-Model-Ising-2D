Pada tugas ini dilakukan simulasi Model Ising 2D untuk mengamati perubahan magnetisasi dan konfigurasi spin pada suhu rendah, suhu kritis, dan suhu tinggi.

1.2 Tujuan

3. Menganalisis transisi fase yang terjadi pada sistem.
4. Membandingkan konfigurasi spin pada suhu rendah, kritis, dan tinggi.

3.1 Data Loading and Time-Domain Visualization
Program simulasi dibuat menggunakan bahasa Python dengan bantuan library NumPy dan Matplotlib. Simulasi menerapkan algoritma Metropolis Monte Carlo pada Model Ising 2D berukuran 20 × 20 spin dengan kondisi awal berupa spin acak (hot start). Program dijalankan sebanyak 200.000 iterasi untuk tiga nilai suhu, yaitu T = 1,0, T = 2,27, dan T = 4,0. Hasil simulasi berupa grafik magnetisasi terhadap iterasi serta visualisasi konfigurasi spin akhir yang digunakan untuk menganalisis perilaku sistem dan transisi fase yang terjadi. 

3.2 Hasil Simulasi T=1.0
	Pada suhu rendah (T = 1,0), energi interaksi antar spin lebih dominan dibandingkan energi termal. Akibatnya, spin-spin cenderung menyusun diri dalam arah yang sama untuk mencapai keadaan energi minimum. Hal ini terlihat dari konfigurasi akhir yang didominasi oleh satu orientasi spin sehingga terbentuk domain magnetik yang besar. Grafik magnetisasi juga menunjukkan nilai yang relatif tinggi dan stabil. Kondisi ini menunjukkan bahwa sistem berada pada fase feromagnetik dengan tingkat keteraturan yang tinggi.

3.3 Hasil Simulasi T=2.27
	Pada suhu kritis (T = 2,27), sistem berada di sekitar titik transisi fase antara feromagnetik dan paramagnetik. Pada kondisi ini terjadi persaingan antara energi interaksi spin dan energi termal sehingga muncul fluktuasi yang cukup besar. Konfigurasi spin menunjukkan terbentuknya domain-domain dengan ukuran yang beragam dan tidak lagi didominasi oleh satu arah orientasi tertentu. Grafik magnetisasi cenderung berfluktuasi lebih kuat dibandingkan suhu rendah. Perilaku ini menunjukkan bahwa sistem sedang mengalami perubahan fase dan sangat sensitif terhadap perubahan suhu.

3.4 Hasil Simulasi T=4.0
	Pada suhu tinggi (T = 4,0), energi termal menjadi lebih dominan dibandingkan interaksi antar spin. Akibatnya, orientasi spin berubah secara acak dan keteraturan sistem berkurang secara signifikan. Konfigurasi akhir menunjukkan pola yang tidak teratur tanpa adanya domain magnetik besar yang terbentuk. Nilai magnetisasi yang diperoleh mendekati nol karena jumlah spin yang mengarah ke atas dan ke bawah relatif seimbang. Hasil ini menunjukkan bahwa sistem berada pada fase paramagnetik, yaitu fase yang ditandai oleh hilangnya magnetisasi spontan akibat pengaruh suhu yang tinggi.

3.5 Analisis Statistik
	Berdasarkan hasil simulasi pada ketiga suhu, terlihat bahwa peningkatan suhu menyebabkan penurunan tingkat keteraturan spin dalam sistem. Pada suhu rendah magnetisasi memiliki nilai terbesar karena spin cenderung sejajar, sedangkan pada suhu tinggi magnetisasi mendekati nol akibat orientasi spin yang acak. Sementara itu, pada suhu kritis terjadi fluktuasi magnetisasi yang paling besar karena sistem berada pada daerah transisi fase. Hasil ini sesuai dengan teori Model Ising 2D yang menyatakan bahwa perubahan suhu dapat mengubah sifat magnetik sistem dari fase feromagnetik menjadi fase paramagnetik.

CASE 1

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# PARAMETER MODEL
# ==========================

N = 20
T = 1.0
J = 1
MC_STEPS = 200000

# ==========================
# HOT START
# ==========================

spins = np.random.choice([-1, 1], size=(N, N))

# ==========================
# FUNGSI DELTA E
# ==========================

def delta_energy(lattice, i, j):

    spin = lattice[i, j]

    neighbors = (
        lattice[(i+1)%N, j] +
        lattice[(i-1)%N, j] +
        lattice[i, (j+1)%N] +
        lattice[i, (j-1)%N]
    )

    dE = 2 * J * spin * neighbors

    return dE

# ==========================
# MAGNETISASI
# ==========================

def magnetization(lattice):
    return np.sum(lattice)/(N*N)

# ==========================
# METROPOLIS
# ==========================

M_history = []

for step in range(MC_STEPS):

    i = np.random.randint(0, N)
    j = np.random.randint(0, N)

    dE = delta_energy(spins, i, j)

    if dE <= 0:
        spins[i, j] *= -1

    else:
        r = np.random.rand()

        if r < np.exp(-dE/T):
            spins[i, j] *= -1

    if step % 100 == 0:
        M_history.append(magnetization(spins))

# ==========================
# PLOT MAGNETISASI
# ==========================

plt.figure(figsize=(8,4))

plt.plot(M_history)

plt.xlabel("Sampling ke-100 langkah")
plt.ylabel("Magnetisasi")

plt.title(f"Magnetisasi vs Waktu (T={T})")

plt.grid()

plt.show()

# ==========================
# KONFIGURASI AKHIR
# ==========================

plt.figure(figsize=(6,6))

plt.imshow(spins, cmap='binary')

plt.title(f"Konfigurasi Akhir (T={T})")

plt.colorbar()

plt.show()

print("Magnetisasi akhir =", magnetization(spins))
```
